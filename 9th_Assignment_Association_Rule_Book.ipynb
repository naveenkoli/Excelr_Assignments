```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
book = pd.read_csv("C:\\Users\\hp\\Downloads\\book.csv")
book
```

## 1. Association rules with 10% Support and 50% confidence

```python
dataset1 = apriori(book, min_support=0.1, use_colnames=True)
dataset1.shape
```

```python
rule1 = association_rules(dataset1 , metric='confidence', min_threshold=0.5 )
```

```python
rule1.shape
```

```python
# as we know the lift ratio should be greater than 1
```

```python
rule1=rule1[rule1.lift>1]
rule1
```

The number of rules we are getting are 49

```python
rule1[['support','confidence','lift']].hist()
```

```python
sn.scatterplot(rule1.support , rule1.confidence)
```
