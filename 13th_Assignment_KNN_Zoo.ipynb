```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier 
```

```python
data = pd.read_csv("C:\\Users\\hp\\Downloads\\Zoo.csv")
data
```

```python
data1 = data.iloc[:,1:]
data1
```

```python
X = data1.iloc[:,:-1]
Y = data1.iloc[:,-1]
```

```python
k_fold = KFold(n_splits=5)
```

```python
model = KNeighborsClassifier(n_neighbors=10)
```

```python
result = cross_val_score(model, X, Y, cv=k_fold)
```

```python
result.mean()
```

```python
# we will do this with the grid search to get the correct number of k-value
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_values = np.array(range(3,21,2))

param_grid = dict (n_neighbors=k_values)
param_grid
```

```python
model2 = KNeighborsClassifier()

grid = GridSearchCV(estimator=model2, param_grid=param_grid)

grid.fit(X,Y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```
